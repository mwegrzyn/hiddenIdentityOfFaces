# Plot Results in Freesurfer Standard Space (fsaverage)

This notebook is based on normalized data, so the scripts can be re-run with only having the tMaps and the fsaverage from Freesurfer.

### import modules

```python
from surfer import Brain, Surface

from mayavi import mlab
mlab.init_notebook()

import matplotlib.pylab as plt
import seaborn as sns
sns.set_style('dark')
%matplotlib inline
```

### show ventral view of brain

```python
import numpy as np
# crop image (taken from https://github.com/mwaskom/Waskom_CerebCortex_2017/blob/master/dimensionality_reduction.ipynb)
def crop(arr):
    x, y = np.argwhere((arr != 255).any(axis=-1)).T
    return arr[x.min() - 20:x.max() + 20, y.min() - 20:y.max() + 20, :]
```

```python
fig = plt.figure(figsize=(16,8))

################

ax = plt.subplot(1,2,1)

b = Brain('fsaverage', 'both', 'inflated',background='white',size=1000)

for h in ['lh','rh']:
    tMap = '../fsavg_maps/tMap_fsavg_landscapes_%s.nii.gz'%h
    b.add_overlay(tMap,hemi=h,min=3,max=11)
    
b.show_view(dict(azimuth=0, elevation=180,focalpoint=[0, 0,-150]))
ax.imshow( crop(b.screenshot()) ); ax.set_xticks([]);ax.set_yticks([]);
ax.set_title('Faces vs Landscapes',fontsize=20)

################

ax = plt.subplot(1,2,2)

b = Brain('fsaverage', 'both', 'inflated',background='white',size=1000)

for h in ['lh','rh']:
    tMap = '../fsavg_maps/tMap_fsavg_hands_%s.nii.gz'%h
    b.add_overlay(tMap,hemi=h,min=3,max=11)
    
b.show_view(dict(azimuth=0, elevation=180,focalpoint=[0, 0,-150]))
ax.imshow( crop(b.screenshot()) ); ax.set_xticks([]);ax.set_yticks([]);
ax.set_title('Faces vs Hands',fontsize=20)

plt.savefig('../figures/fsAvgResults.png',dpi=300,bbox_inches='tight')
plt.show()
```
