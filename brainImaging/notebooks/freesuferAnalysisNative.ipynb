### preprocessing

preproce-sess:  

- sf: File with list of all sessions
- surface: name of the subject whos structural data we project on and which hemispheres should be processed
- fwhm: level of smoothing
- per-run: motion correction within run only (because here we use only one run)
- fsd: name of the subdirectory where the functional data are stored

```bash
preproc-sess -sf sessids -surface pBrain lhrh -fwhm 5 -per-run -fsd bold
```

### specify design

add onset files

```bash
cp ${projectDir}design/landscapeDesign.par ./Sess01/bold/003/design.par
cat Sess01/bold/003/design.par
```

mkanalysis-sess

- mcextreg: Use movement as nuisance regressor
- event-related: can be used for blocked designs too
- nskip: ignore first n scans (we already removed them)
- nconditions: because there is always a 0 baseline, two conditions equals 1

```bash
hemi="rh"

mkanalysis-sess -analysis design.sm5.${hemi} \
-surface pBrain ${hemi} \
-fwhm 5 \
-per-run \
-fsd bold \
-paradigm design.par \
-event-related \
-spmhrf 0 \
-refeventdur 30 \
-polyfit 2 \
-mcextreg \
-nskip 0 \
-TR 3 \
-nconditions 1
```

### contrast

```bash
mkcontrast-sess -analysis design.sm5.${hemi} -contrast act-vs-rest -a 1 -c 0
```

### estimate

```bash
selxavg3-sess -sf sessids -analysis design.sm5.${hemi}
```
