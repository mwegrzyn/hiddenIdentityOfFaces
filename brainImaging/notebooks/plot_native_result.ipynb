# Plot Results in Native Space

### import modules

```python
from surfer import Brain, Surface

from mayavi import mlab
mlab.init_notebook()

import matplotlib.pylab as plt
import seaborn as sns
sns.set_style('dark')
%matplotlib inline
```

### show ventral view of brain

```python
import numpy as np
# crop image (taken from https://github.com/mwaskom/Waskom_CerebCortex_2017/blob/master/dimensionality_reduction.ipynb)
def crop(arr):
    x, y = np.argwhere((arr != 255).any(axis=-1)).T
    return arr[x.min() - 20:x.max() + 20, y.min() - 20:y.max() + 20, :]
```

```python
fig = plt.figure(figsize=(16,8))

################

ax = plt.subplot(1,2,1)

b = Brain('pBrain', 'both', 'inflated',background='white',size=1000)

for h in ['lh','rh']:
    tMap = '../landscapes/Sess01/bold/design.sm5.%s/act-vs-rest/t.nii.gz'%h
    b.add_overlay(tMap,hemi=h,min=3,max=11)
    
b.show_view(dict(azimuth=-1, elevation=180,focalpoint=[0, 0,-150]))
ax.imshow( crop(b.screenshot()) ); ax.set_xticks([]);ax.set_yticks([]);
ax.set_title('Faces vs Landscapes',fontsize=20)

################

ax = plt.subplot(1,2,2)

b = Brain('pBrain', 'both', 'inflated',background='white',size=1000)

for h in ['lh','rh']:
    tMap = '../hands/Sess01/bold/design.sm5.%s/act-vs-rest/t.nii.gz'%h
    b.add_overlay(tMap,hemi=h,min=3,max=11)
    
b.show_view(dict(azimuth=-1, elevation=180,focalpoint=[0, 0,-150]))
ax.imshow( crop(b.screenshot()) ); ax.set_xticks([]);ax.set_yticks([]);
ax.set_title('Faces vs Hands',fontsize=20)

plt.savefig('../figures/fsNativeResults.png',dpi=300,bbox_inches='tight')
plt.show()
```

### show atlas ROIs

```python
fig = plt.figure(figsize=(16,12))

count=0

for i in ['aparc','aparc.a2009s','aparc.DKTatlas40']:
    
    # get brain
    b = Brain('pBrain', 'both', 'pial',background='white',size=1000)
    # get atlas
    for h in ['lh','rh']:
        b.add_annotation(i,borders=False,hemi=h,remove_existing=False)
    # set view
    b.show_view(dict(azimuth=0, elevation=180,focalpoint=[0, 0,-50]))
    
    # make image
    im = crop(b.screenshot())
    ax = plt.subplot(2,3,count+1)
    ax.imshow(im);
    ax.set_xticks([]);ax.set_yticks([]);
    ax.set_title(i, fontsize=20)
        
    count+=1

plt.show()
```
